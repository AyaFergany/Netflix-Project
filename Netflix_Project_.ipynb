```python
# Set-up libraries needed
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
# Transform categorical feature(s) to numeric
le = LabelEncoder()
netflix_copy.show_id = le.fit_transform(netflix_copy.show_id)
netflix_copy.type  = le.fit_transform(netflix_copy.type )
netflix_copy.title = le.fit_transform(netflix_copy.title)
netflix_copy.director  = le.fit_transform(netflix_copy.director )
netflix_copy.cast = le.fit_transform(netflix_copy.cast )
netflix_copy.country = le.fit_transform(netflix_copy.country)
netflix_copy.date_added = le.fit_transform(netflix_copy.date_added)
netflix_copy.rating  = le.fit_transform(netflix_copy.rating )
netflix_copy.duration  = le.fit_transform(netflix_copy.duration )
netflix_copy.description = le.fit_transform(netflix_copy.description)
netflix_copy.listed_in	 = le.fit_transform(netflix_copy.listed_in)
```

```python
x = netflix_copy.drop('type', axis=1)
y = netflix_copy['type']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.metrics import accuracy_score
#Random Forest
random_forest = RandomForestClassifier(n_estimators=16,random_state= 101)
random_forest.fit(x_train, y_train)
# Predict Output
rf_predicted = random_forest.predict(x_test)
# Compare actual and predicted values
actual_vs_predict = pd.DataFrame({'Actual': y_test,
                                'Prediction': rf_predicted})
actual_vs_predict.sample(12)
```

```python
print('RandomForest_Accuracy:', accuracy_score(y_test,rf_predicted))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

# Build and train model
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
# Apply model to validation data
y_predict = model.predict(x_test)
# Compare actual and predicted values
actual_vs_predict = pd.DataFrame({'Actual': y_test,
                                'Prediction': rf_predicted})
actual_vs_predict.sample(12)
```

```python
# Evaluate model
print('Decision tree:', accuracy_score(y_test,y_predict))
```
